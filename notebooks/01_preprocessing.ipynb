```python
df[["Displacement", "Max Power", "Max Torque"]].sample(10, random_state=42)
```

```python
df["Max Power"] = (
    df["Max Power"]
    .str.extract(r"(\d+\.?\d*)")
    .astype(float)
)

df["Max Torque"] = (
    df["Max Torque"]
    .str.extract(r"(\d+\.?\d*)")
    .astype(float)
)

df[["Max Power", "Max Torque"]].head()
```

```python
measurement_cols = [
    "Wheel Base",
    "Width",
    "Height",
    "Length"
]

for col in measurement_cols:
    df[col] = (
        df[col]
        .str.extract(r"(\d+\.?\d*)")
        .astype(float)
    )
```

```python
median_columns = [
    "Wheel Base",
    "Width",
    "Height",
    "Length",
    "Max Torque",
    "Max Power",
    "Values per Cylinder",
    "Displacement",
    "No of Cylinder",
    "No Door Numbers",
    "Seating Capacity"
]

for column in median_columns:
    df[column]= df[column].fillna(df[column].median())

df["engine_cc"] = df["engine_cc"].fillna(
    df["engine_cc"].mode()[0]
)
```
